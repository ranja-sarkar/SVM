```python
import sqlite3
import pandas as pd
```

# Create database

```python

df0 = pd.read_excel('DP_qual.xlsx')
df1 = pd.read_excel('DP_qual_HT1.xlsx')
df11 = pd.read_excel('DP_qual_HT2.xlsx')
df4 = df0.append([df1, df11], ignore_index = True)

df2 = pd.read_excel('DP_quant.xlsx')

dff1 = pd.read_excel('DP_CCS_01.xlsx')
dff2 = pd.read_excel('DP_CCS_0203.xlsx')
dff3 = pd.read_excel('DP_CCS_0405.xlsx')
dff4 = pd.read_excel('DP_CCS_06.xlsx')
dff5 = pd.read_excel("DP_CCS_08.xlsx")
dff6 = pd.read_excel("DP_CCS_09.xlsx")

df3 = dff1.append([dff2, dff3, dff4, dff5, dff6], ignore_index = True)
```

```python
df2.shape, df3.shape, df4.shape
```

```python
#creates a database, connects if it exists

conn = sqlite3.connect("Polymer_Database.db")
```

```python
#storing tables with names in the database

#datavolume
df4.to_sql('DP_qual', conn, index = False)
df2.to_sql('DP_quant', conn, index = False)
df3.to_sql('DP_CCS', conn, index = False)

#subsets - qual
df1.to_sql('ext', conn, index = False)
df11.to_sql('ext_other chemicals', conn, index = False)
df0.to_sql('int+ext', conn, index = False)

#subsets - CCS
dff1.to_sql('01_CCS', conn, index = False)
dff2.to_sql('0203_CCS', conn, index = False)
dff3.to_sql('0405_CCS', conn, index = False)
dff4.to_sql('06_CCS', conn, index = False)
dff5.to_sql('08_CCS', conn, index = False)
dff6.to_sql('09_CCS', conn, index = False)
```

```python
import dash
import dash_html_components as html
import dash_core_components as dcc
from dash.dependencies import Output, Input, State
import dash_table as dt

import pickle
import pandas as pd

external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']
app = dash.Dash(__name__, external_stylesheets = external_stylesheets)

#colors = {'background': '#111111', 'text': '#7FDBFF'}
colors = {'background': '#FFD500', 'text': '#ED1C24'}

app.layout = html.Div([
    html.H3('Search engine for corrosion data', style = {'color': colors['text']}),
    dcc.Input(value = 'Pentagram', id ='box', type = 'text'),
#    dcc.Input(id = 'box', type = 'text'),
    html.Button(children = 'Submit', id ='button', type = 'submit'),
#    html.Div(id = 'container', children = "Enter a keyword and press submit")
#    dt.DataTable(id = 'output_table'),
    html.Div(id = 'content')
                    ],

#   style = {'columnCount': 1}
    style = {'textAlign': 'center'}
                    )


@app.callback(Output('content', 'children'),
    [Input('button','n_clicks')],
    [State('box', 'value')])


#def update_output(n_clicks, value):
#    return 'The input value was "{}" and the button has been clicked {} times'.format(value, n_clicks)

def search_result(n_clicks, keyword):
    
    filepath = 'C:/Users/Ranja.Sarkar/pentagramdic_81.pkl'
    with open(filepath,'rb') as f:
        search_word = pickle.load(f)
    if n_clicks > 0:
        try:
            rr = search_word[str(keyword)]
            df = pd.DataFrame(rr, columns = ('Document', 'TF-IDF score'))
            df = df.sort_values('TF-IDF score', ascending = False)
#    df = df.reset_index()
#    return html.Div([html.Pre(df.to_csv())]) 
            return html.Div([html.Div(id = 'output_table'), dt.DataTable(id = 'table', data = df.to_dict('records'),
            columns = [{"name": i, "id": i} for i in df.columns], style_cell = {'textAlign': 'left'}, style_header = 
            {'backgroundColor': 'white', 'fontWeight': 'bold'}, style_as_list_view = True)
                    ])
        except Exception as e:
#            return (print(e)) ## works without n_clicks
            return html.Div([html.H6('No such keyword found')])


if __name__ == '__main__':
    app.run_server(debug = False)
```
